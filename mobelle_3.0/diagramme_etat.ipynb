Diagramme d'Ã©tat pour l'administrateur

```python
import matplotlib.pyplot as plt
import networkx as nx

# CrÃ©er un nouveau graphique dirigÃ©
G = nx.DiGraph()

# Ajouter des nÅuds pour chaque Ã©tat
states = {
    'A': 'Initial',
    'B': 'Authentification',
    'C': 'Ajouter du personnel',
    'D': 'Voir tous les personnels',
    'E': 'Ajouter des poubelles',
    'F': 'Voir toutes les poubelles',
    'G': 'Voir les poubelles Ã  vider',
    'H': 'Voir la carte',
    'I': 'Final'
}

for state, label in states.items():
    G.add_node(state, label=label)

# Ajouter des transitions entre les Ã©tats
transitions = [
    ('A', 'B', 'DÃ©marrer'),
    ('B', 'C', 'Authentifier'),
    ('B', 'D', 'Authentifier'),
    ('B', 'E', 'Authentifier'),
    ('B', 'F', 'Authentifier'),
    ('B', 'G', 'Authentifier'),
    ('B', 'H', 'Authentifier'),
    ('C', 'I', 'Personnel ajoutÃ©'),
    ('D', 'I', 'Voir tous les personnels'),
    ('E', 'I', 'Poubelle ajoutÃ©e'),
    ('F', 'I', 'Voir toutes les poubelles'),
    ('G', 'I', 'Voir les poubelles Ã  vider'),
    ('H', 'I', 'Voir la carte')
]

for start, end, label in transitions:
    G.add_edge(start, end, label=label)

# Positionner les nÅuds
pos = {
    'A': (0, 5),
    'B': (2, 5),
    'C': (4, 8),
    'D': (4, 6.5),
    'E': (4, 5),
    'F': (4, 3.5),
    'G': (4, 2),
    'H': (4, 0.5),
    'I': (6, 5)
}

# Dessiner le graphique
plt.figure(figsize=(14, 10))
nx.draw(G, pos, with_labels=True, labels=nx.get_node_attributes(G, 'label'), node_size=3000, node_color='lightblue', font_size=12, font_weight='bold', arrows=True)

# Ajouter les Ã©tiquettes des transitions
edge_labels = nx.get_edge_attributes(G, 'label')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red', font_size=10)

plt.title('Diagramme d\'Ã©tat pour l\'administrateur', fontsize=16)
plt.show()
```

Diagramme d'Ã©tat pour le personnel

```python
import matplotlib.pyplot as plt
import networkx as nx

# CrÃ©ation du graphe
G = nx.DiGraph()

# Ajout des nÅuds pour chaque Ã©tat
G.add_node("Initial", shape="circle", style="filled", color="lightblue")
G.add_node("Authentification", shape="circle", style="filled", color="lightblue")
G.add_node("Voir les poubelles", shape="circle", style="filled", color="lightblue")
G.add_node("Voir les poubelles Ã  vider", shape="circle", style="filled", color="lightblue")
G.add_node("Voir la carte", shape="circle", style="filled", color="lightblue")
G.add_node("Final", shape="circle", style="filled", color="lightblue")

# Ajout des transitions entre les Ã©tats
G.add_edge("Initial", "Authentification", label="DÃ©marrer", fontcolor="red")
G.add_edge("Authentification", "Voir les poubelles", label="Authentifier", fontcolor="red")
G.add_edge("Authentification", "Voir les poubelles Ã  vider", label="Authentifier", fontcolor="red")
G.add_edge("Authentification", "Voir la carte", label="Authentifier", fontcolor="red")
G.add_edge("Voir les poubelles", "Final", label="Voir les poubelles", fontcolor="red")
G.add_edge("Voir les poubelles Ã  vider", "Final", label="Voir les poubelles Ã  vider", fontcolor="red")
G.add_edge("Voir la carte", "Final", label="Voir la carte", fontcolor="red")

# Positionnement des nÅuds pour une meilleure lisibilitÃ©
pos = {
    "Initial": (0, 0),
    "Authentification": (2, 0),
    "Voir les poubelles": (4, 2),
    "Voir les poubelles Ã  vider": (4, -2),
    "Voir la carte": (4, -4),
    "Final": (6, 0)
}

# Dessin du graphe
plt.figure(figsize=(12, 8))
nx.draw(G, pos, with_labels=True, node_size=8200, node_color="lightblue", font_size=10, font_weight="bold", arrowsize=20)
edge_labels = nx.get_edge_attributes(G, 'label')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red', font_size=8)

# Titre du diagramme
plt.title("Diagramme d'Ã©tat pour le personnel")

# Affichage du diagramme
plt.show()
```
