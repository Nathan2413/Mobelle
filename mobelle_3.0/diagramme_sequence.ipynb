Diagramme de s√©quence pour administrateur

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Cr√©er une nouvelle figure et un axe
fig, ax = plt.subplots(figsize=(12, 12))

# D√©finir les limites de l'axe et le titre
ax.set_xlim(0, 100)
ax.set_ylim(0, 100)
ax.set_title("Diagramme de S√©quence pour l'Administrateur", fontsize=16)

# Fonction pour ajouter des lifelines avec le texte en haut
def add_lifeline(x, y_start, y_end, label, color, lw=2):
    ax.plot([x, x], [y_start, y_end], color=color, lw=lw)
    ax.text(x, y_end + 2, label, fontsize=12, ha='center', va='bottom', bbox=dict(boxstyle="round,pad=0.3", fc="white", ec=color))

# Fonction pour ajouter des messages
def add_message(x_start, x_end, y, message, direction='->'):
    if direction == '->':
        ax.annotate(message, xy=(x_end, y), xytext=(x_start, y),
                    arrowprops=dict(arrowstyle="->", lw=1.5, color='blue'),
                    fontsize=10, ha='center', va='center', color='blue')
    else:
        ax.annotate(message, xy=(x_end, y), xytext=(x_start, y),
                    arrowprops=dict(arrowstyle="<-", lw=1.5, color='green'),
                    fontsize=10, ha='center', va='center', color='green')

# Ajouter des lifelines avec plus d'espace
add_lifeline(10, 10, 90, 'Administrateur', 'red', lw=1)
add_lifeline(50, 10, 90, 'Syst√®me', 'grey', lw=1)
add_lifeline(90, 10, 90, 'Base de Donn√©es', 'green', lw=1)

# Ajouter des messages avec plus d'espacement
y_pos = 85
step = 4.5

# Se connecter
add_message(10, 50, y_pos, 'Se Connecter')
y_pos -= step
add_message(50, 10, y_pos, 'Confirmer Identifiants', direction='<-')
y_pos -= step
add_message(50, 90, y_pos, 'V√©rifier Identifiants')
y_pos -= step
add_message(90, 50, y_pos, 'R√©sultat de la V√©rification', direction='<-')
y_pos -= step
add_message(50, 10, y_pos, 'Acc√®s Accord√©', direction='<-')

# Ajouter Personnel
y_pos -= step
add_message(10, 50, y_pos, 'Ajouter Personnel')
y_pos -= step
add_message(50, 90, y_pos, 'Enregistrer Personnel')
y_pos -= step
add_message(90, 50, y_pos, 'Confirmation d\'ajout Personnel', direction='<-')


# Ajouter Poubelle
y_pos -= step
add_message(10, 50, y_pos, 'Ajouter Poubelle')
y_pos -= step
add_message(50, 90, y_pos, 'Enregistrer Poubelle')
y_pos -= step
add_message(90, 50, y_pos, 'Confirmation d\'ajout', direction='<-')

# Voir les Poubelles
y_pos -= step
add_message(10, 50, y_pos, 'Voir les Poubelles')
y_pos -= step
add_message(50, 90, y_pos, 'R√©cup√©rer Donn√©es Poubelles')
y_pos -= step
add_message(90, 50, y_pos, 'Donn√©es Poubelles', direction='<-')

# Voir Carte
y_pos -= step
add_message(10, 50, y_pos, 'Voir Carte')
y_pos -= step
add_message(50, 90, y_pos, 'Afficher Carte')
y_pos -= step
add_message(90, 50, y_pos, 'Carte Affich√©e', direction='<-')

# Masquer les axes
ax.axis('off')

# Afficher le diagramme
plt.show()
```
